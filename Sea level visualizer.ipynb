```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv("epa-sea-level.csv")
```

```python
df
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df["Year"],df["CSIRO Adjusted Sea Level"])
plt.xlabel("Year")
plt.ylabel("CSIRO Adjusted Sea Level")
plt.title("Change in sea level")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df["Year"], df["CSIRO Adjusted Sea Level"])

# Get the slope and y-intercept of the line of best fit
slope, intercept, r_value, p_value, std_err = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])

# Create an array of x-values for the line of best fit (from 1880 to 2050)
x_values = range(1880, 2051)

# Create an array of y-values for the line of best fit
y_values = slope * x_values + intercept

# Add the line of best fit to the scatter plot
plt.plot(x_values, y_values, "r")

# Set the axis labels and title
plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Sea Level Rise")

# Show the plot
plt.show()

# Predict the sea level rise in 2050
predicted_2050 = slope * 2050 + intercept
print("Predicted sea level rise in 2050:", predicted_2050, "inches")
```

```python
recent_data = df[df['Year'] >= 2000]
                 
```

```python
slope, intercept, r_value, p_value, std_err = linregress(recent_data['Year'],recent_data['CSIRO Adjusted Sea Level'])
x=np.arange(1880,2050)
y= intercept + slope*x
plt.figure(figsize=(10,6))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.plot(x,y, 'r', label='Line of best fit')

x_new=np.arange(2000,2050)
y_new= intercept + slope*x_new
plt.plot(x_new,y_new, 'g', label='Line of best fit since 2020')

plt.xlabel("Year")
plt.ylabel('Sea level(in inches)')
plt.title('Rise in Sea Level')

plt.legend()

plt.show()

           
```
